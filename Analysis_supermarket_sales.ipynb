```python
import pandas as pd 
import matplotlib.pyplot as plt
import datetime
df=pd.read_csv("C:/Users/PC/Downloads/supermarket_sales.csv")
yangon=df.groupby('city').get_group('Yangon')
mandalay=df.groupby('city').get_group('Mandalay')
naypyitaw=df.groupby('city').get_group('Naypyitaw')
```

## Lợi nhuận theo mảng sản phẩm quý I

Ở thành phố yangon mảng sản phẩm đem lại lợi nhuận tốt nhất là Home and lifestyle lí do là bởi có được sự ủng hộ lớn từ nhóm khách hàng nữ, member và cả normal nhưng lại chưa hoàn toàn tốt đều  trong các tháng của quý I.
Ở thành phố mandalay mảng sản phẩm đem lại lợi nhuận tốt nhất là Sport and travel do có sự ủng hộ nhiều đến từ cả nhóm khách hàng thường và thành viên.Nếu xét theo từng tháng trong quý I thì chỉ chưa đạt lợi nhuận tốt nhất ở tháng 2.
Ở thành phố naypyitaw mảng sản phẩm đem lại lợi nhuận tốt nhất là Food and beverages do có được sự ủng hộ lớn từ nhóm khách hàng nữ, thường và thành viên.Nếu xét theo lợi nhuận từng tháng thì chỉ đạt vị trí tốt nhất ở tháng 3.

```python
profit_product_yangon=yangon.groupby('product_line').sum().sort_values(by='profit',ascending=False)['profit'].plot.bar()
```

```python
profit_product_mandalay=mandalay.groupby('product_line').sum().sort_values(by='profit',ascending=False)['profit'].plot.bar()
```

```python
profit_product_naypyitaw=naypyitaw.groupby('product_line').sum().sort_values(by='profit',ascending=False)['profit'].plot.bar()
```
